```python
import numpy as np

# ---------- Activation Functions ----------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

# ---------- Input Sequence ----------
X = [1, 2, 3 , 4]

# ---------- Weights ----------
# Forget gate
Wf = 0.5
Uf = 0.1
bf = 0

# Input gate
Wi = 0.6
Ui = 0.2
bi = 0

# Candidate cell
Wc = 0.7
Uc = 0.3
bc = 0

# Output gate
Wo = 0.8
Uo = 0.4
bo = 0

# Output layer
Wy = 4
by = 0

# ---------- Initial States ----------
h = 0
C = 0

print("Initial h:", h)
print("Initial C:", C)
print()

# ---------- LSTM Forward ----------
for t, x in enumerate(X, start=1):

    # Forget Gate
    f = sigmoid(Wf*x + Uf*h + bf)

    # Input Gate
    i = sigmoid(Wi*x + Ui*h + bi)

    # Candidate Cell
    C_tilde = tanh(Wc*x + Uc*h + bc)

    # Cell State Update
    C = f*C + i*C_tilde

    # Output Gate
    o = sigmoid(Wo*x + Uo*h + bo)

    # Hidden State
    h = o * tanh(C)

    print(f"Time Step {t}")
    print("f =", round(f,3))
    print("i =", round(i,3))
    print("C_tilde =", round(C_tilde,3))
    print("C =", round(C,3))
    print("o =", round(o,3))
    print("h =", round(h,3))
    print()

# ---------- Predict Next Value ----------
y_pred = Wy*h + by

print("Final hidden state:", round(h,3))
print("Predicted value:", round(y_pred,3))
```
